## A Quick Demo

The remainder of this notebook is intended to show off some of the software capabilities that you will learn in the coming weeks. 

We will:
* load the samples of an audio file into our notebook
* play the audio signal through the computer speakers
* make a nice plot of the signal

## Load required libraries

The libraries that we need are not part of the core of the Python language; they are modules that we can use to expand Python's capabilities.

Before we can use these modules, we need to import them. You will want to put the following lines at the top of nearly every notebook that you will be creating.

````python
## These lines of code go into nearly every notebook
# set up plots directly in this notebook
%matplotlib inline

# load matplotlib module
import matplotlib.pyplot as plt

# and NumPy, too
import numpy as np
````

### Additional modules

For this demo, we need two additional libraries:
1. the sound is stored in a `.wav` file. A function for reading such files is provided by the `SciPy` library.
  + alternatively, there is a `wave` module for interating with `.wav` files; it is common that there are multiple options in the Python eco-system
2. To play sound from within the notebook, we use the `Audio` method provided by the `IPython.display`module.

````python
# additionally, we need a function from scipy to read .wav sound files
import scipy.io.wavfile as wav

# and a function that plays sound inside the notebook
from IPython.display import Audio
````

### Read samples from file

Using the `read()` function from the module that we loaded above, we can now easily obtain the samples of our music signal.

The read method returns two quantities:
1. the sample rate of the file
2. a vector of samples
  + these samples are 16-bit integers since they are a bit more compact to store
  + we can easily convert them to floating point numbers between -1 and 1

````python
## read audio samples from file
# get the sample rate and the samples
fs, samples_int16 = wav.read("music.wav")
# then convert them to floats between -1 and 1
samples = samples_int16 / 2.**15
````

### Play samples through the speaker

The `Audio()` function we imported above, let's us play the signal we loaded through directly from inside the notebook.

Of course, you need speakers to actually hear the sound.

````python
## play audio through the speaker
Audio(samples, rate=fs, autoplay=True)
````

#### A fun thing to try

What happens, if you change `rate=fs` to
* `rate=fs*2`
* `rate=fs/2`

### Make a plot of the signal

We can illustrate the signal by plotting it as a function of time. The resulting plot shows how the signal value changes with time.

Times when the amplitude of the signal is large (either positive or negative) corresponds to periods when the signal is louder.

Can you associate the (seven) beats you hear in the sounds with features of the plot?

````python
## make a nice plot
# we need a time axis: same length as samples, increment between
# samples is 1/fs
tt = np.arange(len(samples))/fs

plt.plot(tt, samples)
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('$s(t)$')

plt.show()

Audio(samples, rate=fs)
````

Some features of the signal become discernible only when we zoom in on the plot.

Here, we look at the signal for a period of 10 milliseconds and then 2 milliseconds.

We will learn in this class that features that are visible at smaller time scales correspond to higher *frequency* components in the signals.

````python
## Zoom in on one of the beats: period 10 milliseconds
tt = np.arange(len(samples))/fs

plt.plot(tt*1e3, samples)
plt.grid()
plt.xlabel('Time (ms)')
plt.ylabel('$s(t)$')
plt.xlim(0.40*1e3, 0.50*1e3)

plt.show()
````

````python
## Zoom in even more on one of the beats: 2 milliseconds
tt = np.arange(len(samples))/fs

plt.plot(tt*1e3, samples)
plt.grid()
plt.xlabel('Time (ms)')
plt.ylabel('$s(t)$')
plt.xlim(0.41*1e3, 0.43*1e3)

plt.show()
````
